## Importing Libraries

```python
import autogen
from autogen import UserProxyAgent, AssistantAgent, GroupChat, GroupChatManager
```

## Set OpenAI API Key

```python
import os
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

## Initialize the OpenAI Model

```python
OPENAI_API_KEY = os.environ['OPENAI_API_KEY']
config_list_gpt = [
    {
        "model": "gpt-4o-mini",
        "api_key": OPENAI_API_KEY
    }
]
```

```python
gpt4_config = {
    "cache_seed": 42,
    "temperature": 0.0,
    "config_list": config_list_gpt,
    "timeout": 120,
}
```

## Creating a UserProxyAgent

```python
user_proxy = UserProxyAgent(
    name="Admin",
    system_message= "A human admin. Interact with the planner to discuss\
    the plan. Plan execution needs to be approved by this admin.",
    code_execution_config = False
)
```

## Creating an AssistanAgent

```python
engineer = AssistantAgent(
    name="Engineer",
    llm_config=gpt4_config,
    system_message= """
    Engineer. You follow an approved plan. You write Python code to solve tasks.
    Wrap the code in a single Python code block. Do not provide incomplete code.
    Only use shell commands if absolutely necessary.
    If execution fails, fix the error and return the full corrected code.
    """,
)
```

```python
scientist = AssistantAgent(
    name="Scientist",
    llm_config=gpt4_config,
    system_message="""
    You are a biomedical scientist. You follow an approved plan.
    Your task is to categorize research papers based on their abstracts.
    Do not write or suggest code. Provide clear and structured categories.
    """,
)
```

```python
planner = AssistantAgent(
    name="Planner",
    llm_config=gpt4_config,
    system_message="""
    You are a Planner agent.
    1. Propose a clear, step-by-step plan.
    2. Label each step as either [Engineer] or [Scientist].
    3. Wait for feedback from Admin and Critic.
    4. Revise the plan until approved.
    Ensure the plan is structured, concise, and executable.
    """,
)
```

## Creating a UserProxyAgent

```python
executor = UserProxyAgent(
    name="Executor",
    system_message="Executor. Execute the code written by the engineer\
    and report the result.",
    human_input_mode="NEVER",
    code_execution_config={"last_n_messages": 3,
                           "work_dir":"paper",
                           "use_docker": False},
)
```

## Creating an AssistantAgent

```python
critic = AssistantAgent(
    name="Critic",
    llm_config = gpt4_config,
    system_message = """Critic. Double check plan, claims, code from other
    agents and provide feedback. Check whether the plan includes adding
    verifiable info such as source URL.
    """,
)
```

## Creating a GroupChat

```python
groupchat = GroupChat(
    agents = [user_proxy, engineer, scientist, planner, executor, critic],
    messages = [],
    max_round = 10
)
```

## Creating a GroupChatManager

```python
manager = GroupChatManager(groupchat=groupchat, llm_config=gpt4_config,)
```

## Run the Workflow

```python
user_proxy.initiate_chat(manager, message="""
Find papers on LLM applications that enhance human learning from arxiv in the last week.
Use code to retrieve real papers (e.g., arXiv API).
Create a markdown table with categories, paper titles, authors and links.
""",
)
```
